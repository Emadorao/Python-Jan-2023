```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 df_gdp=pd.read_csv('RealGDP08012022.csv')
```

```python
df_gdp.head()
```

```python
df_gdp.reset_index(inplace=True)
```

```python
df_gdp.head()
```

```python
df_gdp.columns
```

```python
df_gdp.rename({'index':'tyear','tyear': 'Period', 'Period': 'Agriculture', 'Agriculture':'CropP','CropProduction': 'Livestock',
       'Livestock':'Forestry','Forestry': 'Fishing', 'Fishing':'Industry','Industry':'Mining','MiningAndQuarrying':'CrudeOil',
       'CrudePetroleumAndNaturalGas':'SolidM','SolidMinerals':'CoalMining','CoalMining':'MetalOres',
       'MetalOres':'Quarrying', 'QuarryingAndOtherMinerals':'Manufacturing', 'Manufacturing':'Electricity',
       'ElectricityGasSteamAndAirCon':'WaterS', 'WaterSupplySewageWaste': 'Construction',
       'Construction':'Services', 'Services':'Trade', 'Trade': 'Accommodation','AccommodationAndFoodServices': 'Transport',
       'Transport':'ICT','InformationAndCommunication':'Utilities', 'Utilities':'Entertainment',
       'ArtsEntertainmentAndRecreation':'FinIns', 'FinanceAndInsurance':'RealEstate','RealEstate':'ScienceTech',
       'ProfessionalScientificAndTechnicalServices':'Administratve',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdmin',
       'PublicAdministration':'Education', 'Education':'HealthSer', 'HumanHealthAndSocialServices':'OtherSer',
       'OtherServices':'TOTAL_GDP', 'TOTAL_GDP':'TotalGdpB', 'TotalGDPatConstantBasicPrices':'NONoilBP',
       'NonOilGDPatConstantBasicPrices':'NetTax', 'NetIndirectTaxesOnProducts':'Totalgdpm',
       'TotalGDPatConstantMarketPrices':'M'}, axis=1, inplace=True)
```

```python
df_gdp.isnull().sum()
```

```python
df_gdp.drop(['Mining','CoalMining','MetalOres','Quarrying','Electricity','WaterS','TotalGdpB','NONoilBP','NetTax',
             'Totalgdpm','M'], axis =1, inplace =True)
```

```python
gdp=df_gdp[df_gdp['Period']=='Annual']
```

```python
plt.style.use('ggplot')
plt.plot(gdp['tyear'], gdp['Agriculture'],ls='--',lw=2.5,marker='x',mec='purple',mfc='purple',markevery=5)
plt.xlabel('Year')
plt.ylabel('Agriculture Gdp in Millions')
plt.title('Nigeria GDP in Millions')
plt.show()
```

```python
plt.style.use('dark_background')
plt.plot(gdp['tyear'], gdp['Agriculture'],ls='--',lw=2.5,marker='x',mec='purple',mfc='purple',markevery=5)
plt.xlabel('Year')
plt.ylabel('Agriculture Gdp in Millions')
plt.title('Nigeria GDP in Millions')
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(gdp['tyear'], gdp['Agriculture'],ls='--',lw=2.5,marker='x',mec='purple',mfc='purple',markevery=5, label='Agric')
plt.plot(gdp['tyear'], gdp['Education'],color='yellow',ls='--',lw=2.5,marker='x',mec='purple',mfc='purple',markevery=5,label='Education')
plt.plot(gdp['tyear'], gdp['ICT'],color='red',ls='--',lw=2.5,marker='x',mec='purple',mfc='purple',markevery=5,label='ICT')
plt.xlabel('Year')
plt.ylabel('Gdp in Millions')
plt.title('Nigeria GDP in Millions')
plt.legend()
plt.show()
```
